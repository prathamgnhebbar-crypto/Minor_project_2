```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Data set for DADS June.csv")

print(df.head())
```

```python
print(df.shape)

print(df.columns)

print(df.info())
```

```python
def convert_date(x):
    formats = [
        "%Y-%m-%d",
        "%d-%b-%y",
        "%d %b %Y",
        "%d/%m/%y"
    ]

    for fmt in formats:
        try:
            return pd.to_datetime(x, format=fmt)
        except:
            continue

    return pd.NaT

df["Date"] = df["Date"].apply(convert_date)

print(df["Date"].isna().sum())
```

```python
df["Amount"] = (
    df["Amount"]
    .astype(str)
    .str.replace("₹","",regex=False)
    .str.replace("Rs.","",regex=False)
    .str.replace(",","",regex=False)
    .str.strip()
)

df["Amount"] = pd.to_numeric(df["Amount"])

print(df["Amount"].head())
```

```python
df["Type"] = (
    df["Type"]
    .astype(str)
    .str.strip()
    .str.upper()
)

df["Type"] = df["Type"].replace({
    "DR":"Debit",
    "CR":"Credit"
})

print(df["Type"].value_counts())
```

```python
print("Duplicates Before:",df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After:",df.duplicated().sum())
```

```python
vendor_dict = {
    "Amazon": ["AMAZON", "AMZN"],
    "Amazon Pay": ["AMAZONPAY"],
    "Swiggy": ["SWIGGY", "BUNDL"],
    "Zomato": ["ZOMATO"],
    "Blinkit": ["BLINKIT", "GROFERS"],
    "Zepto": ["ZEPTO"],
    "Instamart": ["INSTAMART"],
    "BigBasket": ["BIGBASKET", "KIRANAKART"],
    "DMart": ["DMART", "AVENUE SUPERMARTS"],
    "Flipkart": ["FLIPKART", "FKART"],
    "Myntra": ["MYNTRA"],
    "Nykaa": ["NYKAA", "FSN E-COMMERCE"],
    "Uber": ["UBER"],
    "Ola": ["OLA", "ANI TECHNOLOGIES"],
    "Rapido": ["RAPIDO"],
    "BMTC": ["BMTC", "TUMMOC"],
    "Starbucks": ["STARBUCKS"],
    "Third Wave": ["THIRDWAVE", "TWC INDIA"],
    "Cafe Coffee Day": ["CCD", "COFFEE DAY"],
    "Truffles": ["TRUFFLES"],
    "Empire": ["EMPIRE"],
    "Meghana Foods": ["MEGHANA"],
    "Restaurant": ["RESTAURANT"],
    "BookMyShow": ["BMS", "BIGTREE"],
    "Netflix": ["NETFLIX"],
    "Spotify": ["SPOTIFY"],
    "Hotstar": ["HOTSTAR", "STAR INDIA"],
    "Groww": ["GROWW", "NEXTBILLION"],
    "Zerodha": ["ZERODHA"],
    "Salary": ["SALARY"],
    "Rent": ["RENT"],
    "Jio": ["JIO", "JIOFIBER"],
    "Airtel": ["AIRTEL", "BHARTI"],
    "Vi": ["VI POSTPAID", "VI-RECHARGE"],
    "BESCOM": ["BESCOM", "BANGALORE ELEC"],
    "BWSSB": ["BWSSB"],
    "Indian Oil": ["IOC", "INDIAN OIL"],
    "HP Petrol": ["HP PETROL"],
    "BPCL": ["BPCL"],
    "Cash Withdrawal": ["ATM-WDL"],
    "P2P Transfer": ["@OKAXIS", "@OKICICI", "@PAYTM", "@HDFCBANK"]
}
```

```python
df["Vendor"] = df["Description"].apply(extract_vendor)
```

```python
def extract_vendor(description):

    description = str(description).upper()

    for vendor,keywords in vendor_dict.items():

        for keyword in keywords:

            if keyword in description:

                return vendor

    return "Unknown"

df["Vendor"] = df["Description"].apply(extract_vendor)

print(df["Vendor"].value_counts())
```

```python
category_dict.update({
    "Flipkart":"E-Commerce",
    "Myntra":"E-Commerce",
    "Nykaa":"E-Commerce",
    "Amazon Pay":"Payments",

    "BigBasket":"Quick Commerce",
    "DMart":"Groceries",

    "Rapido":"Transport",

    "Restaurant":"Dining",
    "Truffles":"Dining",
    "Empire":"Dining",
    "Meghana Foods":"Dining",

    "Netflix":"Subscription",
    "Spotify":"Subscription",
    "Hotstar":"Subscription",

    "Jio":"Utilities",
    "Airtel":"Utilities",
    "Vi":"Utilities",
    "BESCOM":"Utilities",
    "BWSSB":"Utilities",

    "Indian Oil":"Fuel",
    "HP Petrol":"Fuel",
    "BPCL":"Fuel",

    "Rent":"Housing"
})

df["Category"] = df["Vendor"].map(category_dict).fillna("Others")
```

```python
df["Month"] = df["Date"].dt.month_name()

print(df[["Date","Month"]].head())
```

```python
total_credit = df[df["Type"]=="Credit"]["Amount"].sum()

total_debit = df[df["Type"]=="Debit"]["Amount"].sum()

net_savings = total_credit-total_debit

print("Total Credit :",total_credit)

print("Total Debit :",total_debit)

print("Net Savings :",net_savings)
```

```python
top_categories = (
    df.groupby("Category")["Amount"]
    .sum()
    .sort_values(ascending=False)
)

print(top_categories)
```

```python
top_vendors = (
    df.groupby("Vendor")["Amount"]
    .sum()
    .sort_values(ascending=False)
)

print(top_vendors.head(10))
```

```python
monthly_spending = (
    df.groupby("Month")["Amount"]
    .sum()
)

print(monthly_spending)
```

```python
monthly_category = pd.pivot_table(
    df,
    values="Amount",
    index="Category",
    columns="Month",
    aggfunc="sum",
    fill_value=0
)

print(monthly_category)
```

```python
df["Hour"] = df["Time"].str[:2].astype(int)

print(df[["Time","Hour"]].head())
```

```python
hourly_spending = (
    df.groupby("Hour")["Amount"]
    .sum()
)

print(hourly_spending)
```

```python
late_food = df[
    (df["Category"]=="Food Delivery")
    &
    (df["Hour"]>=21)
]

print(late_food)
```

```python
mean = df.groupby("Category")["Amount"].transform("mean")

std = df.groupby("Category")["Amount"].transform("std")

df["z_score"] = (df["Amount"]-mean)/std

anomalies = df[df["z_score"]>2]

print(anomalies[["Date","Vendor","Category","Amount","z_score"]])
```

```python
df["Hour"] = df["Time"].str[:2].astype(int)

df[["Time","Hour"]].head()
```

```python
def time_slot(hour):

    if 5 <= hour < 12:
        return "Morning"

    elif 12 <= hour < 17:
        return "Afternoon"

    elif 17 <= hour < 21:
        return "Evening"

    else:
        return "Night"

df["Time Slot"] = df["Hour"].apply(time_slot)
```

```python
time_analysis = df.groupby("Time Slot")["Amount"].sum()

print(time_analysis)
```

```python
late_food = df[
    (df["Category"]=="Food Delivery") &
    (df["Hour"]>=21)
]

print(late_food[["Date","Vendor","Amount","Hour"]])
```

```python
food = df[df["Category"]=="Food Delivery"]

late_percent = (len(late_food)/len(food))*100

print("Late Night Food Orders:",round(late_percent,2),"%")
```

```python
mean = df.groupby("Category")["Amount"].transform("mean")

std = df.groupby("Category")["Amount"].transform("std")

df["Z Score"] = (df["Amount"]-mean)/std
```

```python
anomalies = df[df["Z Score"]>2]

anomalies = anomalies.sort_values("Z Score",ascending=False)

anomalies[["Date","Vendor","Category","Amount","Z Score"]]
```

```python
print(anomalies.head(10))
```

```python
category_spend = df.groupby("Category")["Amount"].sum()

total_spend = category_spend.sum()

category_percent = (category_spend/total_spend)*100

print(category_percent)
```

```python
archetypes = []
```

```python
if category_percent.get("E-Commerce",0)>15:
    archetypes.append("THE SHOPAHOLIC")
```

```python
if category_percent.get("Investment",0)>10:
    archetypes.append("THE INVESTOR")
```

```python
with pd.ExcelWriter("SpendDNA_Report.xlsx") as writer:

    # Main cleaned dataset
    df.to_excel(writer, sheet_name="Cleaned Data", index=False)

    # Top Categories
    top_categories.to_frame(name="Amount").to_excel(
        writer,
        sheet_name="Top Categories"
    )

    # Top Vendors
    top_vendors.to_excel(
        writer,
        sheet_name="Top Vendors"
    )

    # Monthly Spending
    monthly_spending.to_frame(name="Amount").to_excel(
        writer,
        sheet_name="Monthly Spending"
    )

    # Monthly Category Pivot
    monthly_category.to_excel(
        writer,
        sheet_name="Monthly Category"
    )

    # Time Analysis
    time_analysis.to_frame(name="Amount").to_excel(
        writer,
        sheet_name="Time Analysis"
    )

    # Anomalies
    anomalies.to_excel(
        writer,
        sheet_name="Anomalies",
        index=False
    )

print("✅ SpendDNA_Report.xlsx created successfully!")
```
